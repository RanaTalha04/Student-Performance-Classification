### 2. Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = load_data(os.path.join("..", "data", "processed", "students_Preprocessed_Data.csv"))
df.head()
```

```python
# Columns to drop due to very low correlation with target

low_corr_cols = ["paid", "schoolsup", "famsup", "activities", "nursery" , "absences", "health" , "goout", "freetime", "traveltime", "internet", "romantic", "famrel"]

df = df.drop(columns=low_corr_cols)
```

```python
# Select numeric columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(12,8))
sns.heatmap(df[numeric_cols].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap of Numeric Features")
plt.savefig(os.path.join("..", "reports", "eda_plots", "correlation_heatmap.png"))
plt.show()
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x=col, hue="pass_fail")
    plt.title(f"{col} vs Pass/Fail")
    plt.xticks(rotation=45)
    plt.savefig(os.path.join("..", "reports", "eda_plots", f"{col}_vs_pass_fail.png"))
    plt.show()
```

```python
for col in numeric_cols:
    if col != "pass_fail":  # pass_fail is our target
        plt.figure(figsize=(6,4))
        sns.boxplot(data=df, x="pass_fail", y=col)
        plt.title(f"{col} vs Pass/Fail")
        plt.savefig(os.path.join("..", "reports", "eda_plots", f"{col}_boxplot_pass_fail.png"))
        plt.show()
```

```python
selected_cols = ["age", "studytime", "failures", "pass_fail"]
sns.pairplot(df[selected_cols], hue="pass_fail")
plt.savefig(os.path.join("..", "reports", "eda_plots", "pairplot_selected_features.png"))
plt.show()
```
